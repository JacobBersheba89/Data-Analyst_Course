Using the `requests` library retrieve the current exchange rates from [https://exchangeratesapi.io/](https://exchangeratesapi.io/) and display them as a list:

```
- <RATE 1> EUR/<CURRENCY 1>
- <RATE 2> EUR/<CURRENCY 2>
...
```

The result should be similar to:
```
- 4.5066 EUR/PLN
- 1.5319 EUR/CAD
- 0.86308 EUR/GBP
...
```


````python
import requests

url = "https://api.exchangeratesapi.io/latest"
api_key = "5e68f2acb2681e3b6bd51234d2e6a329"

params = {
    "access_key": api_key,
    "base": "EUR"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    exchange_data = response.json()

    if "rates" in exchange_data:
        rates = exchange_data["rates"]

        print("Current Exchange Rates (Base: EUR):")
        for currency, rate in rates.items():
            print(f"- {rate:.4f} EUR/{currency}")
    else:
        print("Error: 'rates' key not found in the response JSON.")
else:
    print(f"Failed to retrieve exchange rates. Status code: {response.status_code}")
````
