```python
import requests
import psycopg2

API_KEY = "5052fd190cc4baef6464d573e76fb0c2" 

connection = psycopg2.connect(
    user="postgres",
    password="*********************",
    host="localhost",
    port="5432",
    database="classicmodels"
)

cursor = connection.cursor()

cursor.execute("""SELECT customerName, city, country, state, contactFirstname, contactLastName, phone FROM customers
""")

customers = cursor.fetchall()

def get_weather(city, country):
    base_url = f"http://api.openweathermap.org/data/2.5/weather&appid={API_KEY}"
    params = {
        "q": f"{city}, {country}",
        "appid": API_KEY
    }

    response = requests.get(base_url, params = params)

    if response.status_code == 200:
        return response.json()
    else:
        return None

for customer in customers:
    customer_name = customer[0]
    city = customer[1]
    country = customer[2]
    state = customer[3]
    contact_first_name = customer[4]
    contact_last_name = customer[5]
    phone = customer[6]

    weather_data = get_weather(city, country)

    if weather_data:
        weather_conditions = [w['main'] for w in weather_data['weather']]
        if "Rain" in weather_conditions:
            print(f"{customer_name}: {contact_first_name} {contact_last_name}, {phone}")
        else:
            print(f"{customer_name}")
    else: 
        print(f"Could not retrieve weather data for {city}, {country}")

cursor.close()
connection.close()
```
