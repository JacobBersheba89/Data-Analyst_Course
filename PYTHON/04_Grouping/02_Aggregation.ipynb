# Aggregation


Using data from file **product_prices_cleaned.csv**, aggregate the data for each product by month and determine the statistics: `min, max, median, mean, std` for prices (**value** column):

1. skip the national data in the analysis,
1. directly on the object from `groupby`,
1. write a loop that will calculate these values for individual provinces.

Use the `agg` method, and aggregate the data with the `'product','date'` column to complete the exercise.

```python
import pandas as pd

data = pd.read_csv(
    '/Users/jeladobos/Documents/Learning/Data Analyst/5th session/1st Day/01_Data/product_prices_cleaned.csv',  # path to the file with data
    sep=';',  # column separator
    decimal='.'  # decimal separator
)

data['date'] = pd.to_datetime(data['date'], errors='coerce')

data_filter = data[data['province'] != 'POLAND']

group_data = data_filter.groupby(['product', data_filter['date'].dt.to_period('M'), 'province'])['value'].agg(['min', 'max', 'median', 'mean', 'std'])

provinces = data_filter['province'].unique()

for province in provinces:
    province_data = group_data.xs(province, level='province')
    print(f"Statistics for province: {province}")
    print(province_data)
    print("\n" + "="*50 + "\n")

group_data.to_csv('/Users/jeladobos/Documents/Learning/Data Analyst/5th session/1st Day/01_Data/province_product_price_stats.csv', sep=';', index=True)
```
